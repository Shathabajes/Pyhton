```python
import requests

url = "https://api.coinpaprika.com/v1/tickers"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print("Failed to retrieve data:", response.status_code)
```

```python
import pandas as pd
df=pd.json_normalize(data)
df['Timestamp']=pd.to_datetime('now')
df
```
